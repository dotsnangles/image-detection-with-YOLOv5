```python
import os
from sklearn.model_selection import train_test_split

output_path = '/content/image_detection_mini_prjt/dataset'
img_path = '/content/image_detection_mini_prjt/dataset/images'
lb_path = '/content/image_detection_mini_prjt/dataset/labels'

img_list = os.listdir(img_path)

x = [os.path.join(img_path, fname) for fname in img_list]

y = []
for fname in img_list:
    txt = os.path.join(lb_path, fname[:-4]+'.txt')
    with open(txt, 'rt') as f:
       y.append(f.readline().split(' ')[0])

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, train_size=0.8, random_state=None, shuffle=True, stratify=y)

with open(os.path.join(output_path, 'split_train.txt'), 'w') as f:
  f.write('\n'.join(X_train) + '\n')

with open(os.path.join(output_path, 'split_val.txt'), 'w') as f:
  f.write('\n'.join(X_test) + '\n')
```
